```python
import pandas as pd
import matplotlib.pyplot as p
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\rishy\Desktop\rr\lgdata.csv")
p.scatter(data["SAT"],data["GPA"])
p.show()
```

```python
def cost_function(t0,t1,m,x,y):
    s = 0
    for i in range(m):
        s += (t0 + t1*x[i] - y[i])**2
    return s/(2*m)
```

```python
def gradient_decent_theta(Q0,Q1,m,a,x,y):
    g1 = Q0 - (a * ((m*Q0 + Q1*sum(x) - sum(y))/m))
    g2 = Q1 - (a * ((Q0*sum(x) + Q1*sum((x**2)) -sum(x*y))/m))
    return g1,g2
```

```python
x = np.array(list(data['SAT']))
y = np.array(list(data['GPA']))
cost=[]
cc=[]
i=0
#printcounter=0
t0 = 0
t1 = 0
m=84
a = 1/10**8
cc.append(i)
cost.append(cost_function(t0,t1,m,x,y))
p.figure(figsize=(10,10))
while True:
    #c1 = cost_function(t0,t1,m,x,y)
    (t0,t1) = gradient_decent_theta(t0,t1,m,a,x,y)
    c2 = cost_function(t0,t1,m,x,y)
    p.plot(x,t0+(t1*x),color='blue')
    if c2<0.0230:
        break
    else:
        i += 1
        #printcounter =printcounter+1
        cc.append(i)
        cost.append(c2)
    #if printcounter == 1000:
        #print(i)
        #print(c2)
        #print("********************************")
        #printcounter = 0
```

```python
p.scatter(x,y)
p.plot(x,t0+(t1*x),color='blue')
p.show()
p.figure(figsize=(12,10))
p.plot(cc,cost,'^')
p.show()
```
